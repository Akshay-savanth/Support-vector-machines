```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
forest=pd.read_csv('forestfires.csv')
forest
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
forest["month"]=label_encoder.fit_transform(forest['month'])
forest["day"]=label_encoder.fit_transform(forest['day'])
forest["size_category"]=label_encoder.fit_transform(forest['size_category'])
```

```python
forest
```

```python
array=forest.values
X=array[:,0:30]
Y=array[:,30]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[20,15,10,11],'C':[13,14,17,9,7,2.5]}]
gsv=GridSearchCV(clf,param_grid,cv=15)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
clf=SVC(C=13,gamma=20)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
acc=accuracy_score(y_test,y_pred)*100
print("ACCURACY",acc)
confusion_matrix(y_test,y_pred)
```
